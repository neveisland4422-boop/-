```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
orders = pd.DataFrame({
    'order_id': [f"Order_{i}" for i in range(1001, 1019)],
    'region': [
        '华东', '华北', '华南', '西南', '东北', '华东', '华南', '西北', '华东', '华南', 
        '西南', '东北', '华东', '华南', '西南', '华东', '华南', '西北'
    ],
    'product': [
        '机械键盘', '无线鼠标', '显示器', '扩展坞', '机械键盘', '显示器', '无线鼠标', 
        '显示器', '扩展坞', '机械键盘', '无线鼠标', '扩展坞', '显示器', '机械键盘', 
        '扩展坞', '机械键盘', '无线鼠标', '显示器'
    ],
    'quantity': [2, 1, 1, 5, 2, 4, 3, 1, 1, 5, 2, 4, 1, 1, 2, 2, 3, 1],
    'unit_price': [899, 129, 1299, 399, 899, 1299, 129, 1299, 399, 899, 129, 399, 1299, 899, 399, 1299, 129, 1299],
    'member_level': [
        '金卡', '普通', '银卡', '金卡', '银卡', '普通', '金卡', '银卡', '普通', '金卡', 
        '普通', '银卡', '金卡', '普通', '银卡', '金卡', '银卡', '普通'
    ],
    'coupon_rate': [0.05, 0.00, 0.08, 0.10, 0.05, 0.00, 0.12, 0.05, 0.00, 0.08, 0.10, 0.05, 0.00, 0.08, 0.10, 0.05, 0.08, 0.00],
    'salesperson': [
        '小张', '小赵', '小钱', '小孙', '小李', '小周', '小吴', '小郑', '小王', '小冯', 
        '小陈', '小褚', '小卫', '小蒋', '小沈', '小韩', '小杨', '小朱'
    ]
})
```

任务2

```python
analysis = (
    orders
    .assign(
        gross_amount=lambda df: df["quantity"] * df["unit_price"],

        member_discount=lambda df: np.where(
            df["member_level"] == "金卡",
            0.10,
            np.where(df["member_level"] == "银卡", 0.05, 0.00)
        ),

        payable_amount=lambda df: df["gross_amount"]
        * (1 - df["member_discount"])
        * (1 - df["coupon_rate"]),

        shipping_fee=lambda df: np.where(
            df["payable_amount"] >= 1000,
            0,
            20
        ),

        final_amount=lambda df: df["payable_amount"] + df["shipping_fee"]
    )
    .round({
        "gross_amount": 2,
        "member_discount": 2,
        "payable_amount": 2,
        "shipping_fee": 2,
        "final_amount": 2
    })
)

analysis[
    [
        "order_id",
        "quantity",
        "unit_price",
        "gross_amount",
        "member_level",
        "member_discount",
        "coupon_rate",
        "payable_amount",
        "shipping_fee",
        "final_amount"
    ]
].head(8)
```

analysis 是在原始订单表 orders 的基础上生成的新表，新增了原价金额、会员折扣、应付金额、运费和最终成交金额。

任务3

```python
condition_region = analysis["region"].isin(["华东", "华南"])

condition_amount = analysis["final_amount"] >= 700

condition_quantity_or_member = (
    (analysis["quantity"] >= 2)
    | (analysis["member_level"] == "金卡")
)

mask = condition_region & condition_amount & condition_quantity_or_member

important_orders = (
    analysis
    .loc[
        mask,
        [
            "order_id",
            "region",
            "product",
            "quantity",
            "member_level",
            "final_amount"
        ]
    ]
    .sort_values("final_amount", ascending=False)
)

important_orders
```

重点跟进订单同时满足地区、成交金额和数量或会员等级条件，按 final_amount 降序排列后，金额更高的订单排在前面。
在 Pandas 中，每个条件都是一列 True/False。& 表示并且，| 表示或者，它们用于按行合并多个布尔条件。每个条件两边要加括号，是为了让 Python 先计算单个条件，再组合条件，避免运算顺序出错。

任务4

```python
def add_order_level(df):
    result = df.assign(
        order_level=np.where(
            df["final_amount"] >= 2000,
            "战略订单",
            np.where(df["final_amount"] >= 1000, "重点订单", "普通订单")
        )
    )

    return result
```

```python
leveled_orders = analysis.pipe(add_order_level)

level_count = leveled_orders["order_level"].value_counts()

print(level_count)
```

通过 pipe 调用 add_order_level 函数后，订单被划分为战略订单、重点订单和普通订单。

任务5

```python
region_report = (
    analysis
    .pipe(add_order_level)
    .query("final_amount >= 500")
    .groupby(["region", "order_level"])
    .agg(
        order_count=("order_id", "count"),
        quantity_sum=("quantity", "sum"),
        revenue_sum=("final_amount", "sum"),
        revenue_mean=("final_amount", "mean")
    )
    .reset_index()
    .assign(
        revenue_sum=lambda df: df["revenue_sum"].round(2),
        revenue_mean=lambda df: df["revenue_mean"].round(2)
    )
    .sort_values("revenue_sum", ascending=False)
)

region_report
```

region_report 汇总了不同地区、不同订单等级的订单数量、商品数量、成交总额和平均成交金额。

任务6

```python
salesperson_revenue = (
    analysis
    .groupby("salesperson")["final_amount"]
    .sum()
    .sort_values(ascending=False)
)

top_salesperson = salesperson_revenue.index[0]
top_salesperson_amount = salesperson_revenue.iloc[0]

region_revenue = (
    analysis
    .query("salesperson == @top_salesperson")
    .groupby("region")["final_amount"]
    .sum()
    .sort_values(ascending=False)
)

core_region = region_revenue.index[0]
core_region_amount = region_revenue.iloc[0]

region_ratio = core_region_amount / top_salesperson_amount

print("销售人员：", top_salesperson)
print("总成交金额：", round(top_salesperson_amount, 2))
print("核心地区：", core_region)
print("核心地区成交金额：", round(core_region_amount, 2))
print("地区贡献率：", format(region_ratio, ".2%"))
```

```python
print(
    "业务结论："
    + top_salesperson
    + " 是最终成交金额最高的销售人员，其核心地区是 "
    + core_region
    + "，该地区贡献了其总成交金额的 "
    + format(region_ratio, ".2%")
    + "。"
)
```

本次作业使用 Pandas 完成了电商订单的经营分析。首先通过 shape、columns、Series 和 DataFrame 选择快速理解订单数据；然后基于原始订单表构造 gross_amount、member_discount、payable_amount、shipping_fee 和 final_amount 等结算指标；接着通过布尔筛选找出重点跟进订单，并通过函数和 pipe 给订单划分等级。最后使用 groupby 对地区和订单等级进行汇总，形成 region_report，并进一步分析成交金额最高的销售人员及其核心地区。
